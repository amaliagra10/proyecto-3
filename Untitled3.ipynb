# 2. PROCESAMIENTO DE LENGUAJE NATURAL

A. LEMANTIZACIÓN Y TOKENIZACION

```
# creamos el objeto para lemantizar las palabras (las llevamos a su palabra raiz)
wordnet_lemmatizer = WordNetLemmatizer()

# definimos los stopwords
stopwords = nltk.corpus.stopwords.words('spanish')

#definimos la estructura de cada palabra

def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}

    return tag_dict.get(tag, wordnet.NOUN)
```

```
# simplificamos la oración
titular_list=[]
for titular in data.review_title:
    # Vamos a reemplzar los caracteres que no sean leras por espacios
    titular=re.sub("[^a-zA-Z]"," ",str(titular))
    # Pasamos todo a minúsculas
    titular=titular.lower()
    # Tokenizamos para separar las palabras
    titular=nltk.word_tokenize(titular)
    # Aplicamos el Lemmatizer
    frase_lemma = [wordnet_lemmatizer.lemmatize(w, get_wordnet_pos(w)) for w in titular]
    # Eliminamos las palabras de menos de 3 letras
    titular = [palabra for palabra in titular if len(palabra)>3]
    # Sacamos las Stopwords
    titular = [palabra for palabra in titular if not palabra in stopwords]
    # Por ultimo volvemos a unir el titular
    titular=" ".join(titular)
    #dataset["titular_normalizado"] = titular_list
    titular_list.append(titular)
```

```
data["titular_lemm"] = titular_list
data.tail()
```

```
#nos quedamos solo con la columanda lemantizada y tokenizda y con la columna stars
data_lemm = pd.concat([data.titular_lemm,data.stars],axis=1)
data_lemm.dropna(axis=0,inplace=True)  # Por si quedaron titulares vacios
data_lemm.head()
```

### B. VECTORIZACION

B.1 VECTORIZACION CON Ngram=2

```
#Creamos objeto para Vectorizar nuestra lista de titulos con ngram=2 (2 palabras contiguas) 
# Usaremos solo las 500 palabras con mas frecuencia en todo el corpus para generar los vectores
max_features=500
cou_vec=CountVectorizer(max_features=max_features, ngram_range=(1,2)) 
```

```
#Vectorizamos nuestra lista de titulos
matriz_titulos = cou_vec.fit_transform(lista_titulos)

# Tomamos las palabras
all_words = cou_vec.get_feature_names()
```

```
#Definimos nuestros features y target, dividimos en train y test
x = matriz_titulos.toarray()
y = stars

xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```
#definimos función de matriz de confusión

def confusion(ytest,y_pred):
  names=["stars1","stars2", "stars3","stars4","stars5"]
  cm=confusion_matrix(ytest,y_pred)
  f,ax=plt.subplots(figsize=(5,5))
  sns.heatmap(cm,annot=True,linewidth=.5,linecolor="r",fmt=".0f",ax=ax)
  plt.xlabel("y_pred")
  plt.ylabel("y_true")
  ax.set_xticklabels(names)
  ax.set_yticklabels(names)
  plt.show()
  
  return
```

```
#pasamos nuestra lista de titulos por el RandomForest

rf1 = RandomForestClassifier(n_estimators = 50, random_state=42)
rf1.fit(xtrain,ytrain)
print("acc: ",rf1.score(xtest,ytest))
```

El modelo de Randon Forest con un vectorización n-gram =2 me da un accuracy (precisión) de 39.377%, muy baja 

```
y_pred=rf1.predict(xtest)
confusion(ytest,y_pred)
```
